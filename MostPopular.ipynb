```python
import numpy as np
import pandas as pd
import cv2, matplotlib.pyplot as plt
```

```python
transaction_train = pd.read_csv("transactions_train.csv")
```

```python
df = transaction_train[transaction_train['t_dat']>='2020-09-16']
df.shape
```

```python
top12 = ' 0' + ' 0'.join(df.article_id.value_counts().index.astype('str')[:12])
top12
```

```python
rcmnds_img = top12.split()
```

```python
path = "C:/Users/siyuf/Documents/H&M"
import matplotlib.pyplot as plt
```

```python
def plot_recommendation():
    #print(rcmnds)
    fig = plt.figure(figsize=(20, 10))
    for item, i in zip(rcmnds_img, range(1, 13)):
        sub = str(item)[:3]
        image = path + "/"+ sub + "/"+ item +".jpg"
        image = plt.imread(image)
        fig.add_subplot(2, 6, i)
        plt.imshow(image)
```

```python
plot_recommendation()
```
